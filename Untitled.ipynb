### Clasificación de variables:

<h3><li>Cualitativas: 

```python
variables_cualitativas = [
    'Tipo_Producto', 'Tipo_Empaque', 'Metodo_Pago', 'Modo_Entrega', 
    'Rango_Cantidad'
]
```

<h3><li>Cuantitativas:

```python
variables_cuantitativas = [
    'Cantidad', 'Precio_Venta', 'Precio_Unitario', 'Costo_Unitario', 
    'Costo_Produccion', 'Longitud_mt','Capas', 'Denier'
]
```

<h3><li>Temporales:

```python
variables_temporales = ['Fecha_Venta']
```

### Clasificación por tipo numérico: Discretas , Continuas

```python
variables_numericas = ['Cantidad', 'Precio_Venta', 'Precio_Unitario', 'Costo_Unitario','Costo_Produccion', 'Capas','Denier','Longitud_mt']

# Clasificación  Discretas vs Continuas
discretas = []
continuas = []

for var in variables_numericas:
    unique_vals = df_vista_comprension[var].nunique()
    series = df_vista_comprension[var].dropna()
    
    # Detectar si los valores son esencialmente enteros
    if unique_vals < 30 and (series % 1 == 0).all():
        discretas.append(var)
    else:
        continuas.append(var)
print("Variables Discretas:", discretas)# son para enteros
print("Variables Continuas:", continuas)# son para flotantes
```

### Estadística descriptiva:

<h3><li>Tendencia central y Medidas de dispersión

```python
# Variables numéricas que vamos a analizar
columnas_numericas = ['Cantidad', 'Capas','Longitud_mt','Precio_Unitario', 'Precio_Venta', 'Costo_Unitario', 'Costo_Produccion','Denier']

for column in columnas_numericas:
    if column in df_vista_comprension.columns:
        if pd.api.types.is_numeric_dtype(df_vista_comprension[column]):
            print(f"\n Análisis estadístico de '{column}':")

            # Tendencia central
            media = df_vista_comprension[column].mean()
            mediana = df_vista_comprension[column].median()
            moda = df_vista_comprension[column].mode()

            print(f"  - Media: {media:.2f}")
            print(f"  - Mediana: {mediana:.2f}")
            print(f"  - Moda: {moda.values.tolist()}")


            # Medidas de dispersión
            max_val = df_vista_comprension[column].max()
            min_val = df_vista_comprension[column].min()
            rango = max_val - min_val
            q1, q2, q3 = np.percentile(df_vista_comprension[column].dropna(), [25, 50, 75])
            varianza = df_vista_comprension[column].var()
            desviacion = df_vista_comprension[column].std()

            print(f"  - Mínimo: {min_val}")
            print(f"  - Máximo: {max_val}")
            print(f"  - Rango: {rango}")
            print(f"  - Percentil 25 (Q1): {q1}")
            print(f"  - Percentil 50 (Q2 / Mediana): {q2}")
            print(f"  - Percentil 75 (Q3): {q3}")
            print(f"  - Varianza: {varianza:.2f}")
            print(f"  - Desviación estándar: {desviacion:.2f}")
            print("________________________________________")
    else:
        print(f"La columna '{column}' no se encuentra en el DataFrame.")
```

## Exploracion de los datos 

### Análisis univariado

```python
# Análisis de frecuencia para variables categóricas
for var in variables_cualitativas:
    print(f"Frecuencia de {var}:\n", df_vista_comprension[var].value_counts(), "\n")

# Estadísticas para cuantitativas
print("Resumen estadístico de variables cuantitativas:\n")
print(df_vista_comprension[variables_cuantitativas].describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histograma para variables numéricas
for var in variables_cuantitativas:
    plt.figure(figsize=(10, 4))
    sns.histplot(df_vista_comprension[var], kde=True)
    plt.title(f'Distribución de {var}')
    plt.show()


# Gráficos de barras para cualitativas
for var in variables_cualitativas:
    plt.figure(figsize=(7, 3))
    df_vista_comprension[var].value_counts().plot(kind='bar')
    plt.title(f'Frecuencia de {var}')
    plt.xticks(rotation=45)
    plt.show()
```

### Análisis bivariado

<h3><li>Precio vs cantidad

```python
plt.figure(figsize=(7,4))
sns.boxplot(data=df_vista_comprension, x='Rango_Cantidad', y='Cantidad')
plt.title('Distribución de Cantidad por Rango de Cantidad')
plt.show()
```

<h3><li>Precio vs fecha

```python

# Asegurarse que Fecha_Venta esté en formato datetime
df_vista_comprension['Fecha_Venta'] = pd.to_datetime(df_vista_comprension['Fecha_Venta'])

# Graficar precio a lo largo del tiempo por rango de cantidad
plt.figure(figsize=(12, 5))
sns.scatterplot(
    data=df_vista_comprension, 
    x='Fecha_Venta', 
    y='Precio_Venta', 
    hue='Rango_Cantidad',
    palette='Set2',
    alpha=0.7
)
plt.title('Precio de Venta en el Tiempo por Rango de Cantidad')
plt.xlabel('Fecha de Venta')
plt.ylabel('Precio de Venta')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Asegurar formato de fecha
df_vista_comprension['Fecha_Venta'] = pd.to_datetime(df_vista_comprension['Fecha_Venta'])

# Crear columna 'Mes' (año-mes)
df_vista_comprension['Mes'] = df_vista_comprension['Fecha_Venta'].dt.to_period('M')

# Agrupar por mes y sumar la cantidad vendida
ventas_mensuales = df_vista_comprension.groupby('Mes')['Cantidad'].sum()

# Convertir PeriodIndex a datetime para graficar correctamente
ventas_mensuales.index = ventas_mensuales.index.to_timestamp()

# Gráfico de línea
plt.figure(figsize=(12, 5))
plt.plot(ventas_mensuales.index, ventas_mensuales.values, marker='o', linestyle='-')
plt.title('Tendencia de Cantidad Vendida por Mes')
plt.xlabel('Fecha')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Asegurar formato de fecha
df_vista_comprension['Fecha_Venta'] = pd.to_datetime(df_vista_comprension['Fecha_Venta'])

# Crear columna mensual
df_vista_comprension['Mes'] = df_vista_comprension['Fecha_Venta'].dt.to_period('M')

# Agrupar: total de cantidad por mes y por rango
ventas_rango = df_vista_comprension.groupby(['Mes', 'Rango_Cantidad'])['Cantidad'].sum().unstack(fill_value=0)

# Convertir PeriodIndex a Timestamp
ventas_rango.index = ventas_rango.index.to_timestamp()

# Graficar línea por rango
plt.figure(figsize=(12, 6))
ventas_rango.plot(marker='o', linewidth=2)
plt.title('Tendencia de Cantidad Vendida por Rango de Cantidad')
plt.xlabel('Fecha')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Rango de Cantidad')
plt.tight_layout()
plt.show()
```
